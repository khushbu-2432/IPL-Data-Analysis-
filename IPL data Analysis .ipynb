****One Random Game****

```python
random_match_id = np.random.choice(ipl2024['match_id'].unique())
```

```python
random_match_data = ipl2024[(ipl2024['match_id'] == random_match_id) & 
                             (ipl2024['over'] >= 0.1) & (ipl2024['over'] <= 0.6)]
```

```python
random_match_summary = random_match_data.agg({
    'runs_of_bat':'sum', #Total Runs scored
    'extras':'sum',   #Total extras conceded
    'wicket_type': lambda x: x.notnull().sum() #counts of wickets taken
})
```

```python
plt.figure(figsize=(11, 5))
plt.plot(range(1, 4), [random_match_summary['runs_of_bat']] * 3, marker='o', label='Runs Scored')
plt.plot(range(1, 4), [random_match_summary['extras']] * 3, marker='o', label='Extras Conceded')
plt.plot(range(1, 4), [random_match_summary['wicket_type']] * 3, marker='o', label='Wickets Taken')
plt.title(f'Trend in First Over for Match ID {random_match_id}')
plt.xlabel('Match ID')
plt.ylabel('Count')
plt.xticks(range(1, 4), ['Runs Scored', 'Extras Conceded', 'Wickets Taken'])
plt.legend()
plt.grid(True)
plt.show()
```

For all 20 overs

```python
all_overs_data = ipl2024[(ipl2024['over'] >= 0) &(ipl2024['over'] <= 20)]
```

```python
all_overs_grouped = all_overs_data.groupby(['match_id','over'])
```

```python
print(all_overs_grouped)
```

```python
all_overs_summary = all_overs_grouped.agg({
    'runs_of_bat': 'sum',   
    'extras': 'sum',        
    'wicket_type': lambda x: x.notnull().sum()  
})
```

```python
average_trends_all_overs = all_overs_summary.groupby('over').mean()
```

```python
min_runs_per_over = all_overs_data.groupby('over')['runs_of_bat'].min()
max_runs_per_over = all_overs_data.groupby('over')['runs_of_bat'].max()

average_trends_all_overs['min_runs'] = min_runs_per_over
average_trends_all_overs['max_runs'] = max_runs_per_over
```

```python
average_trends_all_overs
```

```python
plt.figure(figsize=(15, 7))
plt.plot(average_trends_all_overs.index, average_trends_all_overs['runs_of_bat'], marker='o', label='Avg Runs Scored')
plt.plot(average_trends_all_overs.index, average_trends_all_overs['extras'], marker='o', label='Avg Extras')
plt.plot(average_trends_all_overs.index, average_trends_all_overs['wicket_type'], marker='o', label='Avg Wickets Taken')
plt.plot(average_trends_all_overs.index, average_trends_all_overs['min_runs'], linestyle='--', color='gray', label='Min Runs')
plt.plot(average_trends_all_overs.index, average_trends_all_overs['max_runs'], linestyle='--', color='gray', label='Max Runs')

plt.title('Average Trends for All Overs in All Matches')
plt.xlabel('Over')
plt.ylabel('Average Count')
plt.xticks(average_trends_all_overs.index)
plt.legend()
plt.grid(True)
plt.show()
```
